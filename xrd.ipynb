```python
import matplotlib.pyplot as plt
from pymatgen.analysis.diffraction.xrd import *
from pymatgen.analysis.diffraction.core import *
from pymatgen import Structure
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
```

```python
with MPRester("dbETTzBnFOmjGseN") as m:

    structure = m.get_structure_by_material_id("mp-861937")
    Conv_struct = SpacegroupAnalyzer(structure).get_conventional_standard_structure() #Obtenir la maille conventionelle au lieu de la primitive
    xrd = XRDCalculator(1.5406) 
    diff = xrd.get_pattern(Conv_struct)
```

```python
theta = diff.x #Les angles 2theta où sont observés les différents pics
hkl = diff.hkls #Les indices hkls de tous les pics et leur multiplicité
print("\n Premier pic: Intensité=",diff.y[0],", angle 2θ =",theta[0],"°, plan",hkl[0][0]["hkl"])
print("\n Second pic: Intensité=",diff.y[1],", angle 2θ =",theta[1],"°, plan",hkl[1][0]["hkl"])
print("\n Troisième pic: Intensité=",diff.y[2],", angle 2θ =",theta[2],"°, plan",hkl[2][0]["hkl"])
print("\n Ces indices sont sous la forme (h,k,i,l). Nous ne tenons donc pas compte du 3e élément de chaque tuple dans notre cas.")
```
